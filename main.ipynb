```python
from sklearn import datasets
import matplotlib.pyplot as plt
from model.linear_regression import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Define parameters
n_samples = 100  # Number of samples
n_features = 1   # Number of features (independent variables)
noise = 15.0      # Standard deviation of the noise

# Generate synthetic data
X, y = datasets.make_regression(n_samples=n_samples, n_features=n_features, noise=noise, random_state=42)
```

```python
plt.scatter(X, y, color='blue', marker='o', edgecolor='k')
plt.title('Synthetic Data for Linear Regression')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.grid(True)
plt.show()
```

```python
lr = 0.001
n_iters = 1000

model_1000_iter = LinearRegression(lr=lr, n_iters=n_iters)

costs_1000_iter = model_1000_iter.fit(X, y)

y_pred_1000_iter = model_1000_iter.predict(X)
```

```python
# Plot the results
plt.scatter(X, y, color='blue', marker='o', edgecolor='k', label='Data')
plt.plot(X, y_pred_1000_iter, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression on Synthetic Data')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(range(1, len(costs_1000_iter) + 1), costs_1000_iter, marker='o', linestyle='-', color='b')
plt.title('Cost vs. Iteration')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.grid(True)
plt.show()
```

```python
lr = 0.001
n_iters = 2000

model_2000_iter = LinearRegression(lr=lr, n_iters=n_iters)

costs_2000_iter = model_2000_iter.fit(X, y)

y_pred_2000_iter = model_2000_iter.predict(X)
```

```python
# Plot the results
plt.scatter(X, y, color='blue', marker='o', edgecolor='k', label='Data')
plt.plot(X, y_pred_2000_iter, color='red', linewidth=2, label='Regression Line (2000 Iterations)')
plt.plot(X, y_pred_1000_iter, color='orange', linewidth=2, label='Regression Line (1000 Iterations)')
plt.title('Linear Regression on Synthetic Data')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(range(1, len(costs_2000_iter) + 1), costs_2000_iter, marker='o', linestyle='-', color='b')
plt.title('Cost vs. Iteration')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.grid(True)
plt.show()
```

```python
# Now let's compare with sklearn's LinearRegression
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
y_pred_sklearn = model.predict(X)

plt.scatter(X, y, color='blue', marker='o', edgecolor='k', label='Data')
plt.plot(X, y_pred_sklearn, color='green', linewidth=3, label='Sklearn Linear Regression')
plt.plot(X, y_pred_2000_iter, color='red', linewidth=2, label='Custom Regression (2000 Iterations)')
plt.plot(X, y_pred_1000_iter, color='orange', linewidth=2, label='Sklearn Regression (1000 Iterations)')
plt.title('Linear Regression on Synthetic Data')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.grid(True)
plt.show()
```
